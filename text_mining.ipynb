```python
## first extract sentesnces using tokenization
sent = nltk.sent_tokenize(elon)
```

```python
len(sent)
```

```python
sent
```

```python
corpus = []

for i in range(len(sent)):
  txt = re.sub('[^a-zA-Z]',' ',sent[i])
  txt = txt.lower()
  corpus.append(txt)
```

```python
corpus
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=800, height=800, background_color="white").generate(str(corpus))

# Plot the wordcloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
